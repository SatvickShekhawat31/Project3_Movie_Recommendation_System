```python
# Importing the dependencies
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Loading the dataset from a csv file from a pandas dataframe
movies_data = pd.read_csv(r'C:\Users\Asus\Desktop\Projects\movieRec\movies.csv')
```

```python
# Printing first 5 columns of the data frame
movies_data.head()
```

```python
# Number of rows and columns
movies_data.shape
```

```python
# Selecting the relevant features for the recommendation system
selected_features=['genres','keywords','tagline','cast','director']
```

```python
# Replacing Null values with Null string
for features in selected_features:
    movies_data[features] = movies_data[features].fillna('')
```

```python
# Combining all 5 selected features
combined_features = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast'] + ' ' + movies_data['director']
print(combined_features)
```

```python
# Converting the text into feature vector
vectorizer = TfidfVectorizer()
feature_vector = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vector)
```

```python
# Getting the similarirty score using cosine similarity
similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
# Getting the input from the user
input_name = input('Enter the movie name: ')
print(input_name)
```

```python
# Creating a list with al the movies name from the list
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# Finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(input_name, list_of_all_titles)
print(find_close_match)
```

```python
closest_match = find_close_match[0]
print(closest_match)
```

```python
# Finding the index with the movie title
index_of_the_movie = movies_data[movies_data.title == closest_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# Getting a list of the similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# Sorting the movie based on their similarirty score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
# Print the similar movies based on the index
print('Movies suggested for you: ')

i=0
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i<30):
        print(f"{i+1}. {title_from_index}")
        i+=1
```
